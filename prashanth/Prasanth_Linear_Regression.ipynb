```python
import pandas as pd
import random
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline

# getting the data
salary_data =  pd.read_csv("Salary_Data.csv") 

#splitting the data into test and train sets
X_train, X_test, y_train, y_test = train_test_split(salary_data.YearsExperience, salary_data.Salary, test_size=0.2, random_state=20)

#training the model
linear_regressor = LinearRegression()
linear_regressor.fit(X_train.values.reshape(-1,1),y_train)

#predicting the salary for years in the test dataset
Prediction = linear_regressor.predict(X_test.values.reshape(-1,1)) # Scikit Learn expects a 2D array, 1-D array will produce error

#plotting the results
Train_data = plt.scatter(X_train, y_train,color='r') #plotting the training set
Test_data = plt.scatter(X_test, y_test, color='b') #plotting the test set
Regression_line =plt.plot(X_train,linear_regressor.predict(X_train.values.reshape(-1,1)),color='k',label=('dfd')) #plotting the regression line
Predicted_Salary = plt.scatter(X_test,Prediction, color='g') # Plotting the Predicted values
plt.legend((Train_data, Test_data, Predicted_Salary),('Train_set', 'Test_set', 'Predicted Salary'),loc='upper left',fontsize=10)
```
