```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import matplotlib.cm as cmap
%matplotlib inline    

#loading the dataset
data = load_breast_cancer().data
y = load_breast_cancer().target

# Z-normalization
SS = StandardScaler()
X = SS.fit_transform(data)

# Defining the KMeans model and predicting the cluster labels( 2- clusters -> malignant and benign )
km_model = KMeans(n_clusters=2, random_state = 74)
kmeans = km_model.fit_predict(X) # returns the predicted cluster labels

# Kmeans accuracy
print('K-means Clustering accuracy:', round(accuracy_score(y, kmeans),2))

# plotting the clusters
plt.scatter(X[:,0],X[:,1],c=kmeans,cmap='cool')
plt.scatter(km_model.cluster_centers_[:,0],km_model.cluster_centers_[:,1],c='black',s=75, label='Cluster centers')
plt.xlabel('Feature-1', fontsize=18)
plt.ylabel('Feature-2', fontsize=18)
plt.title('K-Means on cancer dataset', size=20)
plt.legend()
```
