# Simple Linear Regression

```python
import pandas as pd
import random
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
salarydata = pd.read_csv("D:/Training/Python/Assignments/veridic_machine_learning/Mounika/Simple Linear Regression/Salary_Data.csv")
```

```python
#Assigning data from Salary_Data to X, Y based on dependent and independent variables.
X = salarydata.iloc[:, :1].values
Y = salarydata.iloc[:, 1].values

#slicing training and testing values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/3, random_state = 0)

regressor = LinearRegression()
regressor.fit(X_train, Y_train)

Y_pred = regressor.predict(X_test)
```

```python
plt.scatter(X_train, Y_train, color = "Green")
plt.plot(X_train, regressor.predict(X_train))
plt.title('Salaries Vs Experience')
plt.xlabel('Experience in Years')
plt.ylabel("Salaries")
```

```python
plt.scatter(X_test, Y_test, color = "orange")
plt.plot(X_test, regressor.predict(X_test))
plt.title('Salaries Vs experience')
plt.xlabel('Years of Experience')
plt.ylabel("Salaries")
```
