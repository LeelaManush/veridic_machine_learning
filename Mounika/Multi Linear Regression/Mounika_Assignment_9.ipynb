```python
winequality = pd.read_csv('D:/Training/Python/Assignments/veridic_machine_learning/Mounika/Multi Linear Regression/winequality-white.csv', sep = ';')
```

```python
X = winequality.iloc[:, :-1]
Y = winequality.iloc[:, -1]
```

```python
X = np.append(arr = np.ones((X.shape[0], 1)),values = X, axis = 1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
pred = regressor.predict(X_test)
```

```python
plt.scatter(Y_test, pred, c = 'green')
```

```python
plt.scatter(Y_test, pred, c = 'orange')
plt.xlabel('Quality')
plt.ylabel('Predicted Quality')
plt.title('Predicted Quality based on Actual Quality')
```
